```python
df['Attrition'] = df['Attrition'].replace(['Yes','No'], [1, 0])
```

```python
mod1 = sm.logit('Attrition ~ Age ', data=df).fit()
mod1.summary()
```

```python
mod1 = sm.logit('Attrition ~ Age + WorkLifeBalance', data=df).fit()
mod1.summary()
```

```python
mod1 = sm.logit('Attrition ~ Age + WorkLifeBalance + YearsAtCompany', data=df).fit()
mod1.summary()
```

```python
#TotalWorkingYears
mod1 = sm.logit('Attrition ~ Age + WorkLifeBalance + YearsAtCompany + TotalWorkingYears', data=df).fit()
mod1.summary()
```

```python
mod1 = sm.logit('Attrition ~ Age + WorkLifeBalance + YearsAtCompany + TotalWorkingYears', data=df).fit()
mod1.summary()
```

```python
## create a dataframe with the variable names, OR's and p-values
# np.exp >> calculate the exponential of a given value

df_odds = pd.DataFrame(np.exp(mod1.params), columns = ['OR'])

## pull the pvalues directly from the model

df_odds["pvalue"] = mod1.pvalues

## create a variable to quickly determine sig

df_odds["Sig"] = np.where(df_odds["pvalue"] <= 0.05, "Sig", "Not Sig")

## print new dataset

df_odds
```
