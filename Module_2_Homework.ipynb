```python
import random as rand

num1 = rand.randint(1, 100)
num2 = rand.randint(1, 100)

def multiply_two(x, y):
  print(f"{x} * {y} = {x*y}")

multiply_two(num1, num2)
```

        3 X 3 = 9
        3 X 4 = 12
